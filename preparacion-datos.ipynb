```python
basic_columns = ['Con', 'Contp', 'PP_name', 'PP', 'Reg', 'AgtId', 'Agt', 'Dat','Status','Lgt', 'N_characters', 'Agtp', 'Stage', 'StageSub', 'ThrdPart', 'OthAgr', 'Loc1ISO', 'Loc2ISO', 'UcdpCon', 'UcdpAgr', 'PamAgr', 'CowWar']
gender_columns = ['GeWom', 'GeMe', 'GeMeOth', 'GeSo', 'GeLgbti', 'GeFa']
hr_columns = ['HrGen', 'EqGen', 'HrDem', 'Prot', 'HrFra', 'HrCp', 'HrNi', 'Hrli', 'HrMob', 'HrMob', 'HrDet', 'Med', 'HrCit']
selected_data = data.loc[:, basic_columns + gender_columns + hr_columns]
```

```python
selected_data.head()
```

**Nota:**
Para los campos que son de las categorías genero y derechos humanos, debemos recalcar que el valor 0 indica la no presencia de este atributo en el acuerdo de paz y mayor a 0 la presencia de este atributo en el acuerdo de paz. Para algunos de estos atributos presentan valores que indican 1, 2 o 3 esto significa la inclusión de una subcategoría especifica a ese atributo o el grado de incidencia de este atributo que tuvo en el acuerdo de paz. Para nuestro estudio nos centraremos en el primer nivel que solo indica la presencia o no de este atributo.

```python
selected_data.info()
```

**De acuerdo al informe anterior vemos que las columnas PP_name y Hrli son nulas así que procederemos a eliminarlas de nuestro dataset, así como el atributo PamAgr solo contiene 33 registros con datos.**

```python
selected_data = selected_data.drop(['PP_name', 'Hrli', 'PamAgr'], axis=1)
selected_data.info()
```

**Como resultado de la reducción de dimesiones hemos pasado de 266 atributos a 38.**

# Tranformaciones

Previo a la generación del conjunto de datos final debemos realizar ciertas transformaciones:

- Atributo Con: Algunos valores presentan mas de un país separados por el carácter ‘/’, en este caso como nuestra intención es representar un mapa de países, transformaremos el valor de este atributo seleccionado el primer país presente y los restantes los colocaremos en un nuevo atributo llamando Con2.

```python
selected_data["Con"]= selected_data["Con"].str.split("/", n = 1, expand = True) 
selected_data.head()
```
